## Basic Statistics

Pandas bietet eine ganze Reihe von eingebauten Funktionen um einfache und komplexe Statistiken zu einem Datensatz auszugeben, darunter `.mean()`, `.median()`, `,mode()`, `.min()` und `.max()`.

- Der Mittelwert (Mean) ist der Durchschnitt, der berechnet wird, indem alle Werte addiert und durch die Anzahl der Werte geteilt werden. 
- Der Median hingegen ist der mittlere Wert in einer sortierten Datenreihe, der die Daten in zwei Hälften teilt – unabhängig von extremen Ausreißern. Sollte es sich um eine gerade Anzahl von Werten handeln so ist der Median das Mittel der beiden mittleren Werte. 
- Mode bezeichnet den häufigsten Wert in einer Zahlenreihe. Es kann auch mehrere Modalwerte geben wenn mehrere Zahlen gleich häufig vorkommen. 
- Min. und Max. bezeichnen den niedrigsten, bzw. höchsten Wert in einer sortierten oder unsortierten Zahlenreihe. 

### Ausgabe anpassen

Standardmäßig zeigt Pandas bei DataFrame.describe() sehr viele Nachkommastellen oder wissenschaftliche Notation (1.234567e+06) an. Das kann unübersichtlich wirken – besonders, wenn man die Zahlen nur überblicken und nicht hochpräzise weiterverarbeiten möchte. Über folgenden command kannst du alle Fließkommazahlen mit genau zwei Nachkommastellen anzeigen. Dadurch wird der Vergleich zwischen Spalten einfacher. 

```python
# .describe() outputs all popular statistics of a data set.
# Disable scientific notation for floats
pd.set_option('display.float_format', '{:.3f}'.format)
summary = data.describe()

summary
```

```python
# Calculate the mean
mean_arbeitslosigkeit = data['Arbeitslosigkeit_%'].mean()

# Calculate the median
median_arbeitslosigkeit = data['Arbeitslosigkeit_%'].median()

# Get the min
min_arbeitslosigkeit = data['Arbeitslosigkeit_%'].min()

# Get the max
max_arbeitslosigkeit = data['Arbeitslosigkeit_%'].max()

# Print the results
print(f"Mean Arbeitslosigkeit(%): {mean_arbeitslosigkeit}")
print(f"Median Arbeitslosigkeit(%): {median_arbeitslosigkeit}")
print(f"Min. Arbeitslosigkeit(%): {min_arbeitslosigkeit}")
print(f"Max. Arbeitslosigkeit(%): {max_arbeitslosigkeit}")
```

```python
# Get index of min and max
idx_min = data['Arbeitslosigkeit_%'].idxmin()
idx_max = data['Arbeitslosigkeit_%'].idxmax()

# Get full rows
row_min = data.loc[idx_min]
row_max = data.loc[idx_max]

# Do something with it
print("Row with min:", row_min["Land"])
print("Row with max:", row_max["Land"])
```

## Daten Filtern

Um einen Pandas Dataframe zu Filtern wird die Bedingung direkt in die eckigen Klammern geschrieben und das Resultat einer neuen Variable zugewiesen.


```python
# filtering for an absolute value
landlocked_countries = data[data["Küstenlänge_km"] == 0]
landlocked_countries.head()
```

```python
# filtering above or below a certain value
live_expectency_mean = data["Lebenserwartung"].mean()
print("Durschnittliche Lebenserwartung: ", live_expectency_mean)

live_expectency_above_avg = data[data["Lebenserwartung"] > live_expectency_mean]
live_expectency_above_avg.head()
```

```python
# filtering within a range, e.g. based on quantiles
q1 = data['Lebenserwartung'].quantile(0.25)
q3 = data['Lebenserwartung'].quantile(0.75)

central_half = data[(data["Lebenserwartung"] > q1) & (data["Lebenserwartung"] < q3)]
central_half.head()
```

```python
# generate a random sample of countries
random_subset = data.sample(n=5, random_state=42)
random_subset
```

```python
# select a subset of countries based on their names
north_africa = [
    "Algeria",
    "Egypt",
    "Libya",
    "Mauritania",
    "Morocco",
    "Sudan",
    "Tunisia",
    "Western Sahara"
]

na_subset = data[data["Land"].isin(north_africa)]
na_subset
```
