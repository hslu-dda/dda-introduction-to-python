## Daten importieren

```python
# Pandas Library importieren
import pandas as pd
data = pd.read_excel("./data/african_countries_dataset.xlsx", sheet_name="Data")
```

```python
# Verifiziere das der Import funktioniert hat
data.head()
```

## Subset Erstellen

```python
# select a subset of countries based on their names
north_africa = [
    "Algeria",
    "Egypt",
    "Libya",
    "Mauritania",
    "Morocco",
    "Sudan",
    "Tunisia",
    "Western Sahara"
]

na_subset = data[data["Land"].isin(north_africa)]
na_subset
```

### Spalten Reduzieren

```python
# select only certain columns
col_selection = ["Land", "Fläche_Total_km2", "Bevölkerung", "Medianalter", "BIP_pro_Kopf_USD", "Bevölkerungswachstum_%"]
na_subset = na_subset[col_selection]

na_subset
```

## Daten Aufbereiten

### Zwischenaufgabe

Bevor wir beginnen Werte zu verändern (bspw. durch Skalierung) ist es ratsam eine Kopie des Dataframes anzulegen. Erstelle eine Kopie von `na_subset` und speichere sie in `new_subset`.

```python
new_subset = na_subset.copy()
```

### Summe und Mittelwerte hinzufügen

```python
# create a copy first
sum_avg_data = new_subset.copy()

# add a row at the end for the sum of each columns values
sum_avg_data.loc["Sum"] = sum_avg_data.sum(numeric_only=True)

# add a row at the end for the avg of each columns values
sum_avg_data.loc["Avg"] = sum_avg_data.mean(numeric_only=True)

sum_avg_data
```

### Werte Skalieren


```python
# Ansatz 1: Die originalen Werte mit den neuen Werten überschreiben
scaled_data_1 = new_subset.copy()

scaled_data_1["Fläche_Total_km2"] = scaled_data_1["Fläche_Total_km2"] / 1000
scaled_data_1
```

```python
# Ansatz 2: originale Werte behalten und neue Spalten hinzufügen. 
# Hier kann auch der Skalierungswert explizit erwähnt werden
scaled_data_2 = new_subset.copy()

scaled_data_2.insert(
    scaled_data_2.columns.get_loc("Fläche_Total_km2") + 1,      # position
    "Fläche_Total_km2_Scaled_1000",                             # new column name
    scaled_data_2["Fläche_Total_km2"] / 1000                    # values based on a scale
)

scaled_data_2
```

### Prozente Berechnen

```python
# create a copy
perc_data = new_subset.copy()

# insert a new column with percentages
perc_data.insert(
    perc_data.columns.get_loc("Fläche_Total_km2") + 1,  # position to add column
    "Fläche_Total_km2_Perc", # new column name
    (perc_data["Fläche_Total_km2"] / perc_data["Fläche_Total_km2"].sum() * 100).round(2)    # percentage of total
)

perc_data
```

#### Zwischenaufgabe

Erstelle eine neue Spalte mit den Prozentwerten für die Bevölkerung, arbeite dabei mit der Variable `perc_data`.

```python
# Deine Lösung: 
perc_data.insert(
    perc_data.columns.get_loc("Bevölkerung") + 1,
    "Bevölkerung_%",
    (perc_data["Bevölkerung"] / perc_data["Bevölkerung"].sum() * 100).round(2)
)
perc_data
```

### Differenzen berechnen

Die Differenz bezeichnet die Abweichung zum Mittelwert

```python
diff_data = new_subset.copy()

# always: location to insert, new column name, calculated value
diff_data.insert(
    diff_data.columns.get_loc("BIP_pro_Kopf_USD") + 1,
    "BIP_pro_Kopf_USD_DIFF",
    (diff_data["BIP_pro_Kopf_USD"] - diff_data["BIP_pro_Kopf_USD"].mean()) 
)

diff_data
```
