## Strings

Ob Strings mit ```''``` oder ```""``` geschrieben werden spielt keine Rolle. Nur mischen ```'"``` geht nicht.

````python
"Hallo Echo"
````

````python
'Hallo Echo'
````

Printet man Strings aus geht das mit ```+``` oder ```,```. Bei letzterem wird ein Leerschlag zwischen die Elemente gefügt



````python
print("Hallo" + "Echo")
print('Hallo', 'Echo')
````

Was man so mit Strings machen kann:

````python
hoelzli = "Ich han es Zündhözli azündt"
hoelzli
````

````python
# Umwandeln in Kleinbuchstaben
hoelzli.lower()
````

````python
# uiui
hoelzli.upper()
````

````python
# Zeichenanzahl (~lenght() in R)
len(hoelzli)
````

````python
# Zählen wie viele male ein Wert innerhalb unseres Strings vorkommt
hoelzli.count("n")
````

````python
# Position zurückgeben wo ein Wert in String zum ersten mal vorkommt
hoelzli.find("n")
````

````python
# String in einzelne Elemente unterteilen wo eine Zeichenfolge vorkommt - gibt Liste zurück
hoelzli.split(' ')
````
