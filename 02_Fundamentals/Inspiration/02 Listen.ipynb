## Listen

Listen enthalten eine Aufzählung anderer Elemente. - Anders als in R können sie unterschiedliche Objekttypen gleichzeitig enthalten.

````python
liste_a = ["alpha", "beta", "gamma"]
liste_a
````

````python
# Element hinzufügen

liste_a.append("delta")

liste_a
````

````python
# wir können Listen sortieren

sorted(liste_a)
````

````python
# damit verändert sich die ursprüngliche Liste nicht:
liste_a
````

````python
# die Methode .sort() hingegen verändert das Objekt

liste_a.sort()
````

````python
liste_a
````

````python
# Liste umgekehrt sortieren

liste_a.sort(reverse=True)
liste_a
````

````python
# Elmente entfernen

liste_a.remove("alpha")
liste_a
````

### Listen generieren

Die Range-Funktion erstellt eine Liste im Bereich zweier Zahlen. Oder ab 0 bis zur Zahl.

````python
# geben wir nur eine Zahl an, zählt Python von 0 bis davor.
list(range(10))
````

````python
# geben wir zwei Zahlen an, zählt Python von der ersten bis eine Zahl vor der zweiten Zahl
list(range(2,8))
````

### Subsetting Lists

````python
venus = """D'w. Nuss vo Bümpliz
Isch schön win es Füür i dr Nacht
Win e Rose im Schnee"""

print(venus)
````

````python
# wir teilen den String in eine Liste auf

venus = venus.split(' ')
````

Mit ```[]``` können wir ein Subset einer Liste anzeigen.

````python

# Ein Element anzeigen
venus[1]
````

Python beginnt mit dem Zählen bei null - wie Javascript auch.

````python
venus[0]
````

````python
#Die ersten drei Elemente anzeigen
venus[0:3]
````

````python
# Das letzte Element anzeigen
venus[-1]
````

````python
# alle Elemente nach Position 4 anzeigen
venus[4:]
````
