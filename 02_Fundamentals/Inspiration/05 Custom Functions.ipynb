# 05 Custom Functions

Funktionen sind Zeilen von Code, die in einem Funktions-Namen abgespeichert werden und später wieder aufgerufen werden können.
Damit die Funktion ein Ergebnis zurückspielt muss mit ```return``` gesagt werden was. <br> Nach return endet die Funktion, Code darunter wird ignoriert.

````python

def add_nums(a, b):
    return a+b
````

````python
add_nums(1,2)
````

````python
def bigger_10(x):
    if x > 10:
        res = "Yes"
    else:
        res = "No"
    return res
````

````python
bigger_10(18)
````
